````python
def splitrows(lines: list):
    return [[int(a) for a in list(row)] for row in lines]

def get_data(filename="data.txt"):
    with open(filename) as file:
        contents = file.read().splitlines()
        return splitrows(contents)

def sample1():
    return splitrows("""89010123
                        78121874
                        87430965
                        96549874
                        45678903
                        32019012
                        01329801
                        10456732
                        """.replace(" ", "").strip().splitlines())

def get_trailheads(data: list):
    th = []
    for i in range(len(data)):
        for j in range(len(data[0])):
            if data[i][j] == 0:
                th.append((i, j))
    return th
````

### Naive Solution

For the naive solution, we can recursively search for 0->9 to determine how many peaks we can reach. The only thing we have to be careful of is that we don't count each peak more than once. This is easy enough; we can simply keep track of the peaks we end up at and remove any duplicates at the end.

````python
def get_neighbors(x, y, target, data):
    # Return the coordinates of any neighbors that contain the target number
    # Keeping in mind the bounds of the matrix
    # Neighbors can be either up, down, left, or right
    max_x = len(data[0]) - 1
    max_y = len(data) - 1
    neighbors = []
    
    # Left
    if x > 0 and data[x-1][y] == target:
        neighbors.append((x-1, y))
    # Right
    if x < max_x and data[x+1][y] == target:
        neighbors.append((x+1, y))
    # Up
    if y > 0 and data[x][y-1] == target:
        neighbors.append((x, y-1))
    # Down
    if y < max_y and data[x][y+1] == target:
        neighbors.append((x, y+1))
    
    return neighbors
````

````python
def solve(data: list):
    trailheads = get_trailheads(data)
    print(f"Trailheads: {trailheads}")

    def search(x, y, target):
        # Recursive function to search the path for a 9
        
        # Base case, target == 9
        if target == 9:
            if data[x][y] == 9:
                print(f"Found a 9 at {x},{y}")
                return x, y
            else:
                return None

        neighbors = get_neighbors(x, y, target+1, data)
        print(f"Currently at a {target} ({x},{y}) and looking for ({target+1}) neighbors. Found: {neighbors}")
        
        if len(neighbors) == 0:
            return None
        
        results = []
        for neighbor in neighbors:
            found = search(neighbor[0], neighbor[1], target+1)
            if found:
                print(f"For neighbor {neighbor}, found: {found}")
                
                if type(found) == list:
                    for f in found:
                        results.append(f)
                else:
                    results.append(found)
        return results
        
    total = []
    scores = []
    
    for trailhead in trailheads:
        score = 0
        paths = search(trailhead[0], trailhead[1], 0)
        print(f"Found path(s) for trailhead {trailhead}: {paths}")
        
        # paths = list(set(paths))
        
        for path in paths:
            score += 1
            total.append(path)
            
        scores.append(score)
    
    return total, scores
    
solution = solve(get_data())
````
